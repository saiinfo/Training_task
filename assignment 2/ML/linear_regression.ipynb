```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df['year'], df['per capita income (US$)'], color='red')
plt.show()
```

```python
year = df.drop('per capita income (US$)',axis='columns')
```

```python
year.head()
```

```python
income = df['per capita income (US$)']
```

```python
reg = linear_model.LinearRegression()
reg.fit(year,income)
```

```python
from sklearn.metrics import r2_score

predictions = reg.predict(year)


r_squared = r2_score(income, predictions)
print(f'R-squared: {r_squared}')
```

```python
import matplotlib.pyplot as plt
import numpy as np

predictions = reg.predict(year)

plt.scatter(year, income, color='blue', label='Actual Data')

plt.plot(year, predictions, color='red', label='Predicted Line')

plt.xlabel('Year')
plt.ylabel('Income')
plt.title('Linear Regression Predicted Line and Data Points')
plt.legend()
plt.show()
```
