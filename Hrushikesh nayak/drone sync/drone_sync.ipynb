```python
class Drone:
    def __init__(self, id):
        self.id = id
        self.parent = None
        self.children = []
        
    def send_message(self, destination_id = None, message=None):
        if message is None:
            message = str(input('Drone 1 (root) "type your message": '))
        
        if destination_id is None:
            destination_id = int(input("Kindly select the drone to which you'd like to send a message: "))
            if destination_id == 1:
                print('Root node cannot receive messages.')
                return
        
        if self.id == destination_id:
            print(f"Message '{message}' received by drone {self.id} from Root node")
            return
        
        for child in self.children:
            message_delivered = child.send_message(destination_id, message)



def drone_sync():
    drone1 = Drone(1)
    drone2 = Drone(2)
    drone3 = Drone(3)
    drone4 = Drone(4)
    drone5 = Drone(5)


    drone1.children = [drone2, drone3]
    drone2.parent = drone1
    drone3.parent = drone1

    drone2.children = [drone4, drone5]
    drone4.parent = drone2
    drone5.parent = drone2

    return drone1

def synchronize_drones(node):
    print(f"Synchronizing drone {node.id}")

    for child in node.children:
        synchronize_drones(child)


root_drone = drone_sync()
synchronize_drones(root_drone)


root_drone.send_message()
```
