```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
```

### Features Description
* Age: age
* Sex: Sex (male, female)
* Fare: Passenger Fare (British pound)
* Parch: Number of Parents/Children Aboard
* Pclass: Passenger Class (1 = 1st; 2 = 2nd; 3 = 3rd)
* SibSp: Number of Siblings/Spouses Aboard
* Title: title of their name (Miss, Mr, Mrs, Master)
* Family_Size: size of the family.
* Embarked: Port of Embarkation (C = Cherbourg; Q = Queenstown; S = Southampton)
* Survival: Survived (0 = No; 1 = Yes)

```python
# Load the dataset
dataset = pd.read_csv("datasets/titanic-data.csv")
dataset.head()
```

```python
dataset.describe(include='all')
```

```python
# dropping certain columns that don't have any impact on the survival rate.
dataset.drop(["passenger_id", "name", "sib_sp", "parch", "ticket", "cabin", "embarked"], axis=1, inplace=True)
dataset.head()
```

```python
# converting the categorical variable to numerical

dataset['gender'] = dataset['gender'].map({'female': 0, 'male': 1}).astype(int)
print(dataset.head)
```

```python
# Checking for null values
dataset.isna().sum()
```

```python
# Filling the null values
dataset['age'] = dataset['age'].fillna(dataset['age'].mean())
```

```python
dataset.isna().sum()
```

```python

# Separating the features and target 
X = dataset.drop(columns=['survived'] ,axis = 1)
y = dataset['survived']
print(X.columns)
```

```python
# Splitting the dataset into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train.shape ,y_train.shape)
print(X_test.shape ,y_test.shape)
```

```python
model = GaussianNB()
model.fit(X_train ,y_train)
```

```python
# Evaluating
y_pred = model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix: ")
print(cm)

print("Accuracy of the Model: {0}%".format(accuracy_score(y_test, y_pred)*100))
```
