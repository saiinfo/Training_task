```python
#Step 1: Import Libraries
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,mean_squared_error
```

```python
#Step 2:Load CSV
data=pd.read_csv('car_data.csv')
```

```python
data
```

```python
data=data.dropna()
```

```python
data.shape
```

```python
data.replace({'Fuel_Type':{'Petrol':0, 'Diesel':1, 'CNG':2 }},inplace=True)
```

```python
data.replace({'Seller_Type':{'Dealer':0, 'Individual':1 }},inplace=True)
```

```python
data.replace({'Transmission':{'Manual':0, 'Automatic':1 }},inplace=True)
```

```python
data
```

```python
#Feature Selection
X = data[['Year','Kms_Driven','Fuel_Type','Selling_Price','Seller_Type','Transmission','Owner']]
y = data['Present_Price']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Initialize the model
model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model
model.fit(X_train, y_train)
```

```python
X_test
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
single_row=[[2016,10980,0,10.11,0,0	,0]]


prediction = model.predict(single_row)

print("Predicted Present Price:", prediction[0])
```

```python
new_row=[[2018,11980,0,13.11,0,0,0]]


prediction = model.predict(new_row)

print("Predicted Price: For New Row ", prediction[0])
```

```python
mse = mean_squared_error(y_test, y_pred)

rmse = np.sqrt(mse) 

r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"RMSE: {rmse}")
print(f"R^2 Score: {r2}")
```
