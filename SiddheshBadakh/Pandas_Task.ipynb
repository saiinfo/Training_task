Pandas is a Python library used for working with data sets.

It has functions for analyzing, cleaning, exploring, and manipulating data.

Pandas provides two primary data structures that are widely used for data manipulation: Series and DataFrame.

1. Series
Description: A Series is a one-dimensional labeled array capable of holding data of any type (integers, strings, floats, etc.). It is similar to a column in a spreadsheet or a 1D array in NumPy.

2. DataFrame
Description: A DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous data structure with labeled axes (rows and columns). Think of it as a table (like an Excel spreadsheet) where each column is a Series.
